```python
from ultralytics import YOLO
import torch
import os

# Environment Information


print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Available : {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print(f"GPU : {torch.cuda.get_device_name(0)}")


# Create Output Directory

# All training runs, checkpoints, plots, logs, and
# exported models will be stored inside this directory.


PROJECT_DIR = "/content/drive/MyDrive/PathHoleDetection/training_runs"

os.makedirs(PROJECT_DIR, exist_ok=True)


# Load Pretrained YOLO26 Model

# Using YOLO26 Nano version as the starting point.
# Transfer learning allows the model to converge faster
# and usually provides better accuracy than training
# from scratch.


model = YOLO("yolo26n.pt")


# Model Training


results = model.train(


    # Dataset Configuration


    # Path to dataset YAML file.
    # The YAML file contains:
    #   - train image path
    #   - validation image path
    #   - class names
    #   - number of classes
    data="/content/drive/MyDrive/PathHoleDetection/PotholeDetection_data/data.yaml",


    # Core Training Parameters


    # Total number of passes through the training dataset.
    # Higher values allow more learning but can increase
    # risk of overfitting.
    epochs=100,

    # Input image size.
    # All images are resized to 640x640 before training.
    # Larger values improve small-object detection but
    # increase GPU memory usage.
    imgsz=640,

    # Number of images processed in one iteration.
    # Increase if GPU memory allows.
    batch=8,

    # GPU device to use.
    # device=0 means first GPU.
    device=0,

    # Number of CPU workers used to load images.
    # Higher values speed up data loading.
    workers=8,


    # Optimizer Configuration

    # AdamW optimizer generally provides stable training
    # and good convergence.
    optimizer="AdamW",

    # Final learning rate factor.
    # Final LR = lr0 × lrf.
    lr0=0.001,
    lrf=0.01,

    # L2 regularization.
    # Helps reduce overfitting.
    weight_decay=0.001,

    # Momentum term.
    # Helps stabilize optimization and accelerates
    # convergence.
    momentum=0.937,


    # Warmup Configuration


    # Number of warmup epochs.
    # Gradually increases learning rate during the
    # first few epochs for stable training.
    warmup_epochs=3,

    # Initial momentum during warmup.
    warmup_momentum=0.8,

    # Initial learning rate for bias parameters.
    warmup_bias_lr=0.1,

    # Data Augmentation

    # Enable augmentation pipeline.
    augment=True,

    # Random rotation range in degrees.
    # Images can rotate between -5 and +5 degrees.
    degrees=5,

    # Random horizontal/vertical image shift.
    # 0.1 = 10% translation.
    translate=0.1,

    # Random zoom in/out.
    # 0.2 allows ±20% scaling.
    scale=0.2,

    # Random shear transformation.
    # Simulates slight perspective distortions.
    shear=2.0,

    # Horizontal flip probability.
    # 0.5 means 50% of images are flipped.
    fliplr=0.5,

    # Vertical flip probability.
    # Disabled because roads are rarely upside down.
    flipud=0.0,

    # Mosaic augmentation probability.
    # Combines four images into one training sample.
    # Helps improve generalization.
    mosaic=0.5,

    # MixUp augmentation probability.
    # Combines two images together.
    # Disabled for pothole detection.
    mixup=0.0,

    # Copy-Paste augmentation probability.
    # Mainly useful for segmentation tasks.
    copy_paste=0.0,

    # HSV Hue augmentation.
    hsv_h=0.015,

    # HSV Saturation augmentation.
    hsv_s=0.4,

    # HSV Brightness augmentation.
    hsv_v=0.4,


    # Validation Settings


    # Run validation after every epoch.
    val=True,

    # Early stopping patience.
    # Training stops if validation metric does not
    # improve for 20 consecutive epochs.
    patience=20,

    # Checkpoint & Logging Settings


    # Save model checkpoints.
    save=True,

    # Save checkpoint every 5 epochs.
    save_period=5,

    # Generate training plots such as:
    # loss curves, PR curves, confusion matrix, etc.
    plots=True,


    # Reproducibility


    # Random seed.
    # Ensures consistent results between runs.
    seed=42,

    # Enable deterministic operations.
    # Improves reproducibility but may slightly reduce
    # training speed.
    deterministic=True,


    # Performance Optimization


    # Cache dataset in memory.
    # Speeds up training if RAM allows.
    cache=True,

    # Automatic Mixed Precision (FP16).
    # Reduces GPU memory usage and increases speed.
    amp=True,


    # Output Directory Settings


    # Parent directory where all runs are stored.
    project=PROJECT_DIR,

    # Name of this training run.
    # Results will be saved under:
    # PROJECT_DIR/yolo26n_production
    name="yolo26n_production",

    # Overwrite existing run folder if it exists.
    exist_ok=True
)

# Training Summary


print("\nTraining completed successfully.")

# Directory where this training run is stored.
print(f"\nRun directory:\n{results.save_dir}")

# Path to the best-performing checkpoint selected
# using validation metrics during training.
BEST_MODEL_PATH = os.path.join(
    results.save_dir,
    "weights",
    "best.pt"
)

print(f"\nBest model path:\n{BEST_MODEL_PATH}")
```

```python
from ultralytics import YOLO
import json
import cv2

# Load Model

model = YOLO("/content/drive/MyDrive/PathHoleDetection/training_runs/yolo26n_production/weights/best.pt")


# Image Path


image_path = "/content/drive/MyDrive/PathHoleDetection/PotholeDetection_data/test/images/021_png.rf.RQrwzMt6SBHX6EUQUYPh.png"


# Run Inference

results = model.predict(
    source=image_path,
    conf=0.25,
    save=False,
    verbose=False
)


# Get Image Dimensions

img = cv2.imread(image_path)
height, width = img.shape[:2]


# Build JSON Response

response = {
    "image": image_path,
    "width": width,
    "height": height,
    "num_detections": 0,
    "detections": []
}

for result in results:

    boxes = result.boxes

    for box in boxes:

        x1, y1, x2, y2 = map(float, box.xyxy[0])

        confidence = float(box.conf[0])

        class_id = int(box.cls[0])

        class_name = model.names[class_id]

        detection = {
            "class_id": class_id,
            "class_name": class_name,
            "confidence": round(confidence, 4),
            "bbox": {
                "xmin": round(x1, 2),
                "ymin": round(y1, 2),
                "xmax": round(x2, 2),
                "ymax": round(y2, 2)
            }
        }

        response["detections"].append(detection)

response["num_detections"] = len(response["detections"])


# Print JSON

print(json.dumps(response, indent=4))
```

Model Performence On Unseen (Test) Data

```python
from ultralytics import YOLO
import pandas as pd
import cv2
import os
from glob import glob


# Load Trained Model


MODEL_PATH = "/content/drive/MyDrive/PathHoleDetection/training_runs/yolo26n_production/weights/best.pt"

model = YOLO(MODEL_PATH)


# Test Image Directory


TEST_DIR = "/content/drive/MyDrive/PathHoleDetection/PotholeDetection_data/test/images"


# Output Excel Path


OUTPUT_EXCEL = "/content/drive/MyDrive/PathHoleDetection/test_predictions.xlsx"


# Supported Image Extensions


image_files = []

for ext in ["*.jpg", "*.jpeg", "*.png", "*.bmp"]:
    image_files.extend(glob(os.path.join(TEST_DIR, ext)))

print(f"Total Test Images Found: {len(image_files)}")


# Store Results


all_results = []


# Run Inference


for image_path in image_files:

    try:

        img = cv2.imread(image_path)

        if img is None:
            print(f"Skipping unreadable image: {image_path}")
            continue

        height, width = img.shape[:2]

        results = model.predict(
            source=image_path,
            conf=0.25,
            verbose=False,
            save=True
        )

        detection_count = 0

        for result in results:

            boxes = result.boxes

            if len(boxes) == 0:

                all_results.append({
                    "image_name": os.path.basename(image_path),
                    "image_path": image_path,
                    "width": width,
                    "height": height,
                    "num_detections": 0,
                    "predicted_class": "No Detection",
                    "confidence": 0,
                    "xmin": "",
                    "ymin": "",
                    "xmax": "",
                    "ymax": "",
                    "status": "No Pothole Detected"
                })

            else:

                for box in boxes:

                    detection_count += 1

                    x1, y1, x2, y2 = map(float, box.xyxy[0])

                    confidence = float(box.conf[0])

                    class_id = int(box.cls[0])

                    class_name = model.names[class_id]

                    all_results.append({
                        "image_name": os.path.basename(image_path),
                        "image_path": image_path,
                        "width": width,
                        "height": height,
                        "num_detections": len(boxes),
                        "predicted_class": class_name,
                        "confidence": round(confidence, 4),
                        "xmin": round(x1, 2),
                        "ymin": round(y1, 2),
                        "xmax": round(x2, 2),
                        "ymax": round(y2, 2),
                        "status": "Pothole Detected"
                    })

    except Exception as e:

        print(f"Error processing {image_path}")
        print(e)


# Save Results to Excel


df = pd.DataFrame(all_results)

df.to_excel(
    OUTPUT_EXCEL,
    index=False
)

print("\n===================================")
print("Inference Completed Successfully")
print("===================================")
print(f"Total Images Processed : {len(image_files)}")
print(f"Excel Saved To        : {OUTPUT_EXCEL}")
print("===================================")

# Preview first few rows

df.head()
```

ACTUAL ACCURACY COMPARED WITH GROUND TRUTH

```python
from ultralytics import YOLO
import pandas as pd


# Load Best Model


MODEL_PATH = "/content/drive/MyDrive/PathHoleDetection/training_runs/yolo26n_production/weights/best.pt"

model = YOLO(MODEL_PATH)


# Evaluate On Test Dataset

metrics = model.val(
    data="/content/drive/MyDrive/PathHoleDetection/PotholeDetection_data/data.yaml",
    split="test",
    conf=0.25,
    iou=0.5,
    plots=True,
    save_json=True
)


# Print Metrics


precision = metrics.box.mp
recall = metrics.box.mr
map50 = metrics.box.map50
map5095 = metrics.box.map

f1 = (
    2 * precision * recall / (precision + recall)
    if (precision + recall) > 0
    else 0
)

print("\n" + "=" * 60)
print("YOLO26 TEST DATASET EVALUATION")
print("=" * 60)

print(f"Precision      : {precision:.4f}")
print(f"Recall         : {recall:.4f}")
print(f"F1 Score       : {f1:.4f}")
print(f"mAP@50         : {map50:.4f}")
print(f"mAP@50-95      : {map5095:.4f}")

print("=" * 60)
```
